```python
# 내 풀이
def solution(heights):
    a = []
    while heights:
        if len(heights) == 1:
            a.append(0)
            break

        c = heights.pop()

        for i in range(len(heights)-1,-1,-1):            
            if c < heights[i]:
                a.append(i+1)
                break
            if i == 0:
                a.append(0)

    a.reverse()
    
    return a
```

```python
solution([1,5,3,6,7,6,5])
```

```python
#다른사람풀이. 깔끔하긴한데 이중포문이라 느려지지 않을까?
def solution2(h):
    ans = [0] * len(h)
    for i in range(len(h)-1, 0, -1):
        for j in range(i-1, -1, -1):
            if h[i] < h[j]:
                ans[i] = j+1
                break
    return ans
```

```python
import random
a = [random.randint(1,1000) for i in range(100000)]
```

```python
%time
solution(a)
```

```python
%time
solution2(a)
```
