```python
%time
# 진짜 풀이 ( 와 정답... 그런데 시간이 오래걸림 ㅠ)
def solution(s):
    
    if len(s) < 3:
        return len(s)   
    #길이별로나누기
    a = []
    for i in range(1,(len(s)//2)+1):
        b = []
        for j in range(len(s)):
            if i*j+i <= len(s)+i:
                b.append(s[i*j:i*j+i])

        while "" in b:
            b.remove("")
        a.append(b)

    #겹치는것들 1추가후 리스트화
    d = []
    for i in range(len(a)):
        c = ""
        for j in range(len(a[i])):

            if j+1 == len(a[i]):
                c += a[i][j]

            elif a[i][j] == a[i][j+1]:
                c += "1"
            else:
                c += a[i][j]
        d.append(c)
    d = list(map(list,d))

    num = []
    for i in range(100):
        num.append(i)

    #1들 제거, current =0변경, next += current, 0제거, string화
    num = [str(i) for i in range(1001)]

    for i in range(len(d)):
        for j in range(len(d[i])):
            if d[i][j] in num and d[i][j+1] in num:
                d[i][j+1] = str(int(d[i][j]) + int(d[i][j+1]))
                d[i][j] = str(0)

        while str(0) in d[i]:
            d[i].remove(str(0))

    for i in range(len(d)):
        for j in range(len(d[i])):
            if d[i][j] in num:
                d[i][j] = str(int(d[i][j])+1)

    for i in range(len(d)):
        d[i] = "".join(d[i])
    #리턴
    return min([len(i) for i in d])
```

```python
# 다른사람 풀이인데 한번 해보자
def compress(text, tok_len):
    words = [text[i:i+tok_len] for i in range(0, len(text), tok_len)]
    res = []
    cur_word = words[0]
    cur_cnt = 1
    for a, b in zip(words, words[1:] + ['']):
        if a == b:
            cur_cnt += 1
        else:
            res.append([cur_word, cur_cnt])
            cur_word = b
            cur_cnt = 1
    return sum(len(word) + (len(str(cnt)) if cnt > 1 else 0) for word, cnt in res)

def solution(text):
    return min(compress(text, tok_len) for tok_len in list(range(1, int(len(text)/2) + 1)) + [len(text)])

a = [
    "aabbaccc",
    "ababcdcdababcdcd",
    "abcabcdede",
    "abcabcabcabcdededededede",
    "xababcdcdababcdcd",

    'aaaaaa',
]

for x in a:
    print(solution(x))
```
