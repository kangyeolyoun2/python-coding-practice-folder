```python
class BinaryTree:
    
    def __init__(self):
        self.__root = None
        self.__node = None
    
    ## Root 설정
    @property
    def root(self):
        return self.__root
    
    @root.setter
    def root(self, root):
        self.__root = root
    
    
    ## 노드 만들기
    def make_node(self, data):
        if data == ".":
            return None
        
        node = Node()
        node.data = data
        return node
    
    ## 노드 찾기 (preorder로 찾지 뭐...)
    def find_node(self, data):
        cur = self.root
        self.__node = None
        self.__find_node(data, cur)
    
    
    def __find_node(self, data, cur):
        if not cur: # 자식 노드가 없으면
            return
        
        if self.__node: # 노드를 찾았으면 나머지 빠르게 종료
            return
        
        if data == cur.data: # 데이터를 찾으면 객체에 그 노드 저장
            self.__node = cur
            return

        self.__find_node(data, cur.left) # 못찾으면 재귀
        self.__find_node(data, cur.right) # 못찾으면 재귀
        
    def make_tree(self, data, left, right):
        if not self.root:
            self.root = self.make_node(data)
        
        self.find_node(data)
        cur = self.__node
        l = self.make_node(left)
        r = self.make_node(right)
        cur.left, cur.right = l, r

    def preorder_traverse(self, cur, func):
        if not cur:
            return
        
        func(cur.data)
        self.preorder_traverse(cur.left, func)
        self.preorder_traverse(cur.right, func)
        
    def inorder_traverse(self, cur, func):
        if not cur:
            return
        
        self.inorder_traverse(cur.left, func)
        func(cur.data)
        self.inorder_traverse(cur.right, func)
        
    def postorder_traverse(self, cur, func):
        if not cur:
            return
        
        self.postorder_traverse(cur.left, func)
        self.postorder_traverse(cur.right, func)
        func(cur.data)
        
        
N = int(input())
bt = BinaryTree()

for _ in range(N):
    parent, left, right = input().split()
    bt.make_tree(parent, left, right)

func = lambda x: print(x, end="")
bt.preorder_traverse(bt.root, func)
print()
bt.inorder_traverse(bt.root, func)
print()
bt.postorder_traverse(bt.root, func)
```
